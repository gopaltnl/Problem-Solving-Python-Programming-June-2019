```python
# Voltage vs Time
import numpy as np
t = [i/10 for i in range(1,21)]
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.grid()
plt.title('Time Vs Voltage Graph')
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.show()
```

```python

# Voltage vs Time
import numpy as np
t = [i/10 for i in range(1,21)]
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.plot(t,v,'bo')
plt.plot(v,t,'ro')
plt.grid()
plt.title('Time Vs Voltage Graph')
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.show()
plt.savefig('images/voltage')
```

```python
# plot income arkansas from 2005 to 2013

import pandas as pd
df=pd.read_csv('datafiles/income.csv')
years=df.columns[2:]
years=list(range(2005,2014))
years
incomearacances=df.values[3,2:]
incomearacances
plt.title('Income Of Akansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomearacances)
plt.grid()
plt.show()
```

```python
# Income of All states from 2005 to 2013
incomeallstates= []

for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)

plt.title("ALL STATES INCOME")
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid()
plt.show()
```
